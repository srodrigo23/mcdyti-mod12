# Exploratory Data Analysis (EDA)

This notebook is used for performing exploratory data analysis on the dataset containing information about institutions of higher education in the US. The analysis includes data visualization and initial insights.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set visualisation style
sns.set(style='whitegrid')

# Load the dataset
data_path = '../../data/Most-Recent-Cohorts-Scorecard-Elements.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Display basic information about the dataset
df.info()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the distribution of key numerical features
plt.figure(figsize=(12, 6))
sns.histplot(df['C150_4_POOLED_SUPP'], bins=30, kde=True)
plt.title('Distribution of Completion Rate')
plt.xlabel('Completion Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the relationship between completion rate and median earnings
plt.figure(figsize=(12, 6))
sns.scatterplot(x='C150_4_POOLED_SUPP', y='MD_EARN_WNE_P10', data=df)
plt.title('Completion Rate vs Median Earnings')
plt.xlabel('Completion Rate')
plt.ylabel('Median Earnings 10 Years After Entry')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

## Initial Insights

1. The dataset contains various features related to institutions of higher education, including completion rates and median earnings.
2. There are missing values in several columns that need to be addressed before modeling.
3. Visualizations indicate potential relationships between completion rates and median earnings, which may be explored further in modeling.
